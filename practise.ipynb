## Cleaning the dataset


```python
# setting the index to the s45 column.
bribes.set_index("s45")
```

```python
# replacing the asterix with 0

bribes = bribes.replace(to_replace=' *** ',value='0')
```

```python
# defining a function for removing the commas in the numerical variable columns.

def remove_comma(data):
    for col in data:
        data[col]=data[col].str.replace(',','')
    return data
```

```python
remove_comma(bribes)
```
